```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

En la carpeta Dataset, nos encontramos con 4 archivos: credits.csv, movies_dataset(cargado el 5/6/23),movies_dataset(cargado el 7/6/23) y movies_dataset.csv

Procederemos a cargar estas bases de datos y compararlas para elegir con cual trabajar.

```python
movies_raw_csv = pd.read_csv('data/movies_dataset.csv',low_memory=False)
movies_raw_csv.info()
movies_raw_csv.head(3)
```

```python
movies_raw_xls = pd.read_excel('data/movies_dataset.xlsx')
movies_raw_xls.info()
movies_raw_xls.head(3)
```

```python
movies_raw_xls1 = pd.read_excel('data/movies_dataset(1).xlsx')
movies_raw_xls1.info()
movies_raw_xls1.head(3)
```

```python
# Nulos
nulos_csv = (
    movies_raw_csv
    .isnull()
    .sum()
    .sum()
)

nulos_xls = (
    movies_raw_xls
    .isnull()
    .sum()
    .sum()
)

nulos_xls1 = (
    movies_raw_xls1
    .isnull()
    .sum()
    .sum()
)

# Duplicados

duplicados_csv =(
    movies_raw_csv
    .duplicated(keep=False)
    .sum()
)

duplicados_xls =(
    movies_raw_xls
    .duplicated(keep=False)
    .sum()
)

duplicados_xls1 =(
    movies_raw_xls1
    .duplicated(keep=False)
    .sum()
)

print('Tamaño\n'
      'csv:  ', movies_raw_csv.shape,'\n'
      'xls:  ', movies_raw_xls.shape,'\n'
      'xls1: ', movies_raw_xls1.shape, '\n')

print( 'Valores nulos\n'
      'Los valores nulos del archivo csv son:  ', nulos_csv ,'\n'
      'Los valores nulos del archivo xls son:  ', nulos_xls ,'\n'
      'Los valores nulos del archivo xls1 son: ', nulos_xls1 ,'\n')


print( 'Filas duplicadas\n'
      'Las filas duplicadas del archivo csv son:  ', duplicados_csv ,'\n'
      'Las filas duplicadas del archivo xls son:  ', duplicados_xls ,'\n'
      'Las filas duplicadas del archivo xls1 son: ', duplicados_xls1 ,)
```

Podemos ver que los tres archivos poseen el mismo tamaño y la misma cantidad de filas duplicadas pero el archivo csv tiene menos valores nulos que los xls.

¿Cuáles son las columnas que tienen datos el en archivo csv y nulos en los excel?

```python
columnas_diferentes_xls_xls = []
for columna in movies_raw_xls.columns:
    if movies_raw_xls1[columna].isnull().sum() != movies_raw_xls[columna].isnull().sum():
        columnas_diferentes_xls_xls.append(columna)
print('Las columnas con datos diferentes entre los archivos xls son: ',columnas_diferentes_xls_xls)

columnas_diferentes_csv_xls = []
for columna in movies_raw_csv.columns:
    if movies_raw_csv[columna].isnull().sum() != movies_raw_xls[columna].isnull().sum():
        columnas_diferentes_csv_xls.append(columna)
print('Las columnas con datos diferentes entre los archivos csv y xls son: ',columnas_diferentes_csv_xls)

columnas_diferentes_csv_xls1 = []
for columna in movies_raw_csv.columns:
    if movies_raw_csv[columna].isnull().sum() != movies_raw_xls1[columna].isnull().sum():
        columnas_diferentes_csv_xls.append(columna)
print('Las columnas con datos diferentes entre los archivos csv y xls1 son: ',columnas_diferentes_csv_xls1)
```

Podemos ver que los archivos xls poseen la misma cantidad de datos nulos pero difieren del archivo csv en las columnas 'overview' y 'tagline'.

```python
#Creacion de un dataset con las columnas 'overview' y 'tagline' de los archivos xls y csv para compararlos.
df = pd.concat([movies_raw_csv[['overview','tagline']], movies_raw_xls[['overview','tagline']]], axis=1)
df.columns=['overview_csv','tagline_csv','overview_xls','tagline_xls']
df.isna().sum()
```

El archivo csv tiene menos cantidad de valores nulos (solo 6 menos).

```python
# Busqueda de los datos nulos que no coinciden en la columna 'overview'
condicion1 = (~df['overview_csv'].isna()) & (df['overview_xls'].isna())
df.loc[condicion1, ['overview_csv', 'overview_xls']]
```

```python
condicion2 = (df['overview_csv'].isna()) & (~df['overview_xls'].isna())
df.loc[condicion2, ['overview_csv', 'overview_xls']]
```

```python
# Busqueda de los datos nulos que no coinciden en la columna 'tagline'
condicion3 = (~df['tagline_csv'].isna()) & (df['tagline_xls'].isna())
df.loc[condicion3, ['tagline_csv', 'tagline_xls']]
```

```python
# Busqueda de los datos nulos que no coinciden en la columna 'tagline'
condicion4 = (df['tagline_csv'].isna()) & (~df['tagline_xls'].isna())
df.loc[condicion4, ['tagline_csv', 'tagline_xls']]
```

Podemos ver que las filas que poseen datos nulos en el archivo csv poseen un espacio vacio en el csv por lo que ambos archivos son iguales.

```python
# Eliminamos los valores nulos para comparar.
df = df.dropna()
df[['overview_csv','overview_xls']][df['overview_csv'] != df['overview_xls']]
```

```python
df[['tagline_csv','tagline_xls']][df['tagline_csv'] != df['tagline_xls']]
```

Hemos encontrado que el archivo xls tiene ausencia de comillas y que eso lo diferencia del csv. Por lo que, para continuar esta etapa utilizaremos solamente el archivo 'movies_dataset.csv'
